```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each class
classes = ['W', 'N1', 'N2', 'N3', 'R']
precision = [0.88, 0.57, 0.95, 0.87, 0.75]
recall = [0.90, 0.49, 0.86, 0.97, 0.89]
f1_score = [0.89, 0.52, 0.90, 0.92, 0.81]

x = np.arange(len(classes))  # X-axis for each class
width = 0.3  # Width of the candlestick body

# Function to plot candlestick with body (precision vs recall) and wicks (f1-score)
def plot_candlestick(ax, open_data, close_data, high_data, low_data, label):
    for i in range(len(classes)):
        # Define open (precision/recall) and close for the body
        open_value = open_data[i]
        close_value = close_data[i]

        # High and Low are the extremes among precision, recall, and F1-score
        high_value = high_data[i]
        low_value = low_data[i]

        # Body of the candlestick (rectangle): between open and close
        color = 'green' if close_value >= open_value else 'red'
        ax.bar(x[i], abs(close_value - open_value), bottom=min(open_value, close_value), width=width, color=color, edgecolor='black')

        # Plot the wicks: from low to high
        ax.plot([x[i], x[i]], [low_value, high_value], color='black', lw=2)
    
    # Set the labels
    ax.set_xticks(x)
    ax.set_xticklabels(classes)
    ax.set_ylabel(label)
    ax.set_title(f'{label} Candlestick Chart')

# Create a figure with 3 subplots (one for precision, recall, and F1-score)
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Plot for Precision (precision vs recall, with f1-score as the wick)
plot_candlestick(axs[0], precision, recall, f1_score, f1_score, 'Precision')

# Plot for Recall (recall vs precision, with f1-score as the wick)
plot_candlestick(axs[1], recall, precision, f1_score, f1_score, 'Recall')

# Plot for F1-Score (f1-score vs precision, with recall as the wick)
plot_candlestick(axs[2], f1_score, recall, precision, f1_score, 'F1-Score')

# Add some space between subplots
plt.tight_layout()

# Display the plots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Extracted values from your data
classes = ['W', 'N1', 'N2', 'N3', 'R']
precision = [0.89, 0.63, 0.93, 0.86, 0.70]
recall = [0.86, 0.25, 0.86, 0.88, 0.78]
f1_score = [0.86, 0.29, 0.85, 0.87, 0.74]

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.25
index = np.arange(len(classes))

# Create shadow bars with a darker color
shadow_offset = 0.02  # Offset for the shadow
shadow_color = 'gray'

# Plot shadow bars for each metric
ax.bar(index - shadow_offset, [h + 0.05 for h in precision], bar_width, color=shadow_color, alpha=0.5)
ax.bar(index + bar_width - shadow_offset, [h + 0.05 for h in recall], bar_width, color=shadow_color, alpha=0.5)
ax.bar(index + 2 * bar_width - shadow_offset, [h + 0.05 for h in f1_score], bar_width, color=shadow_color, alpha=0.5)

# Plot original bars for each metric
bar1 = ax.bar(index, precision, bar_width, label='Precision', color='b')
bar2 = ax.bar(index + bar_width, recall, bar_width, label='Recall', color='g')
bar3 = ax.bar(index + 2 * bar_width, f1_score, bar_width, label='F1 Score', color='r')

# Add aesthetics
ax.set_xlabel('Sleep Stages')
ax.set_ylabel('Scores')
ax.set_title('Precision, Recall, and F1 Score for Each Class')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(classes)

# Change legend location to right outside the chart
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Helper function to attach a text label above each bar
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Annotate original bars
autolabel(bar1)
autolabel(bar2)
autolabel(bar3)

plt.tight_layout()
plt.show()
```
